```python
def get_population(size=10000,mean=50,SD=15):
    """ 
    size=10000,mean=50,SD=15

    """

    numbers=[]
    # size=10000
    for i in range(size):
        number=random.gauss(mean,SD)
        numbers.append(number)
    return numbers
get_population()
```

```python
def get_sample(size=100):

    size_of_sample=size
    sample=[]
    for i in range(size_of_sample):
        number=random.choice(numbers)
        sample.append(number)
    return sample


get_sample()
```

```python
def get_sample_statistics(sample=sample):
    sample_statistics={}
    mean=sum(sample)/len(sample)
    mean_with_np=numpy.mean(sample)
    SD=numpy.std(sample)
    
    N=len(sample)
    correction_factor=(N/(N-1))**0.5
    SD_plus=SD*correction_factor

    sample_statistics['mean']=mean
    sample_statistics['mean_with_np']=mean_with_np
    sample_statistics['SD']=SD
    sample_statistics['SD_plus']=SD_plus
    return sample_statistics
get_sample_statistics()
```

```python
def get_sample_of_sample_means(how_many,sample_size):
    sample_means=[]
    for i in range(how_many):
        sample=get_sample(sample_size)
        sample_means.append(sum(sample)/(len(sample)))
    return get_sample_statistics(sample_means)
get_sample_of_sample_means(100,5)
```

```python
def get_sample_statistics_from_small_sample(sample=sample):
    sample_statistics={}
    mean=sum(sample)/len(sample)
    mean_with_np=numpy.mean(sample)
    SD=numpy.std(sample)
    
    N=len(sample)
    correction_factor=(N/(N-1))**0.5
    SD_plus=SD*correction_factor

    sample_statistics['mean']=mean
    sample_statistics['mean_with_np']=mean_with_np
    sample_statistics['SD']=SD
    sample_statistics['SD_plus']=SD_plus
    sample_statistics['SE_for_z']=SD*(N**0.5)/N
    sample_statistics['SE_for_t']=SD_plus*(N**0.5)/N
    return sample_statistics
sample=[78,83,68,72,88]
get_sample_statistics_from_small_sample(sample)
```

```python
# z_test=(observed-expected)"/SE_for_z_test
# t_test=(observed-expected)"/SE_for_t_test
sample=[78,83,68,72,88]
result=get_sample_statistics_from_small_sample(sample)
expected=70
observed=result['mean']
SE_for_t_test=result['SE_for_t']
dof=len(sample)-1
t_test=(observed-expected)/SE_for_t_test
t_test,dof
```
